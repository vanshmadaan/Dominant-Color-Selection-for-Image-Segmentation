### Dominant Color Extraction for Image Segmentation

- Image Segmentation


-  Segmentation partitions an image into regions having similar visual appearance corresponding to parts of objects


-  We will try to extract the most dominant 'K' Colors using K-Means


-  We can apply K-Means with each pixel will reassigned to the closest of the K Colors, leading to segmentation

```python
#We will group together regions which are visually similar or having similar colors and we are going to do it for any given value of K (K=2 means we want to represent our image using only two colors).
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv2.imread('elephant.jpg')    #Read an image into BGR format
im = cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
original_shape = im.shape
print(im.shape)
```

```python
plt.imshow(im)    #Reads as RGB
plt.show()
```

```python
#We will reshape/flatten this image into a linear array. So, we will flatten each channel of the image

all_pixels = im.reshape((330*500,3))    #Instead of using 330*500, we can use -1 and thus it will automatically figure out what should be the shape when you are making it linear
print(all_pixels.shape)   

#Here,we can see that there are 165000 features, each having 3 channels. So, for each channel we are having one linear array
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors = 4

km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixels)

#What is happenening is, there is a 3d color space(RGB) and there are a lot of colors inside it and we are trying to find 4 dominant colors and we are trying to cluster all the colors into 4 type of colors and each pixel will be allotted one color according to the mean of the cluster to which it is nearest to.
```

```python
centers = km.cluster_centers_     #These cluster centers are the RGB values of the colors which K-Means has. These are the most dominant colors.
```

```python
centers = np.array(centers,dtype='uint8')    #Here, we use unsigned int because color values can only be positive and we are taking them as 8 bit integers beause we just want to represent the numbers between 0 to 255 and to represent these numbers 8 bit integer is sufficient.
```

```python
print(centers)    #These are RGB values for 1,2,3 and 4th colors
```

### Plot what all colors are these -


```python
#We are going to make one sub-plot. So, we are going to make a (1x4,i) subplot. In the ith partition, we are going to draw the ith color.

i = 1

plt.figure(0,figsize=(8,2))
colors = []

for each_col in centers:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    
    #Color Swatch -
    a = np.zeros((100,100,3),dtype='uint8')    #Here, we are creating one matrix and lets say we want this matix to be of greenish color. So, it will be a (100,100,3) matrix and we are going to fill it with green color and then we are going to display this matrix.
    a[:,:,:] = each_col
    plt.imshow(a)
    
plt.show()    
```

### Segmenting Our Original Image -

```python
#We will map every pixel with some value
```

```python
new_img = np.zeros((330*500,3),dtype='uint8')   #New image having same shape as the original flattened image. It has 3 channels and all the pixels have been laid in one row

#Here, we are going to iterate over every pixel and we are going to see what is the closest center to this pixel.
#So, from the K-Means labels, we are going to assign each pixel one new color and this color will be out of those 4 colors.

print(new_img.shape)
```

```python
colors
```

```python
km.labels_     
#We can see that it is a matrix and for each pixel it is giving me some id (0,1,2 or 3). So, we are going to pick colors of that id.
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]    #For every integer predicted, we are going to find what is the mapped color
    
new_img = new_img.reshape((original_shape))   #We will reshape it into the original shape. Here, we give the original shape so that we can reshape our matrix to an (wxhx3) image
plt.imshow(new_img)
plt.show()

#If you want to extract elephants and grass, then we can simply find out those pixel values which have this color id.m
```
